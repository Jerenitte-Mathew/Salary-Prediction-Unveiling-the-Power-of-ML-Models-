```python
x=df["YearsExperience"].values
y=df["Salary"].values
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
x_train
```

```python
y_train
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
x_train=x_train.reshape(-1,1)
x_test=x_test.reshape(-1,1)
```

```python
x_train.shape
```

#### Linear Regression model

```python
from sklearn.linear_model import LinearRegression
lmregressor=LinearRegression()
lmregressor.fit(x_train,y_train)
```

```python
y_pred=lmregressor.predict(x_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import r2_score
r2_lr=r2_score(y_test,y_pred)
print("r2 score for Linear Regression model: ",r2_lr)
```

```python
from sklearn import metrics
mae_lr=metrics.mean_absolute_error(y_test,y_pred)
print("Mean Absolute Error: ",mae_lr)
mse_lr=metrics.mean_squared_error(y_test,y_pred)
print("Mean Squared Error: ",mse_lr)
rmse_lr=np.sqrt(metrics.mean_squared_error(y_test,y_pred))
print("Root Mean Squared Error: ",rmse_lr)
```

```python
evaluation_table=[]
evaluation_table.append(["Linear Regression", r2_lr, mae_lr, mse_lr, rmse_lr])
print(evaluation_table)
```

```python
# visualization of Tested vs Predicted

plt.scatter(x_test,y_test,color="orange",label="Actual") # original value
plt.scatter(x_test,lmregressor.predict(x_test),color="green",label="Predicted") # predicted value
plt.plot(x_test, lmregressor.predict(x_test), color="blue", linestyle="dashed")
plt.title("Test vs predict")
plt.legend()
plt.show()
```

```python
# visualization of training test

plt.scatter(x_train,y_train,color="red") # original value
plt.scatter(x_train,lmregressor.predict(x_train),color="green") # predicted value
plt.plot(x_train, lmregressor.predict(x_train), color="blue", linestyle="dashed", label="Regression Line")
plt.title("year of experience vs Salary")
plt.xlabel("year of experience")
plt.ylabel("Salary")
plt.legend()
plt.show()
```

#### Decision tree Regressor

#### Support Vector Machine (SVM) Regressor

```python
# from sklearn.preprocessing import StandardScaler
# scaler = StandardScaler()
# x_train_scaled = scaler.fit_transform(x_train)
# x_test_scaled = scaler.transform(x_test)


from sklearn.svm import SVR
svmregressor=SVR(kernel='rbf')
svmregressor.fit(x_train,y_train) #_scaled
```

```python
y_pred=svmregressor.predict(x_test) #_scaled)
y_pred
```

```python
y_test
```

```python
r2_svm=metrics.r2_score(y_test,y_pred)
print("r2 score for SVM: ",r2_svm)
```

```python
mae_svm=metrics.mean_absolute_error(y_test,y_pred)
print("Mean Absolute Error: ",mae_svm)
mse_svm=metrics.mean_squared_error(y_test,y_pred)
print("Mean Squared Error: ",mse_svm)
rmse_svm=np.sqrt(metrics.mean_squared_error(y_test,y_pred))
print("Root Mean Squared Error: ",rmse_svm)
```

```python
evaluation_table.append(["SVM Regressor", r2_svm, mae_svm, mse_svm, rmse_svm])
# print(evaluation_table)
```

```python
plt.scatter(x_test,y_test,color="orange",label="Actual") # original value
plt.scatter(x_test,svmregressor.predict(x_test),color="green",label="Predicted") # predicted value
plt.plot(x_test, svmregressor.predict(x_test), color="blue", linestyle="dashed",label="Regression Line")
plt.legend()
plt.show()
```

```python
# Convert results to DataFrame
model_evaluation_table = pd.DataFrame(evaluation_table, columns=["Model", "R² Score", "MAE", "MSE", "RMSE"])

print(model_evaluation_table)
```
